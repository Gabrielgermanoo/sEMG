# Preparando o ambiente

```python
import pandas as pd
from ydata_profiling import ProfileReport
from scipy.io import loadmat
import os
```

```python
path = '../data/S1'
data = loadmat(f'{path}/S1_E1_A1.mat')
print(data.keys())
```

# Analisando as features presentes no Dataset - Ninapro para o DB1 - S1 - S1_E1_A1

```python
for key in data.keys():
    if key.startswith('__'):  # Ignore meta keys
        continue
    print(f"{key}: {type(data[key])}, shape: {data[key].shape}")
```

```python
df = pd.DataFrame(data['emg'])  # Convert 'emg' to DataFrame (10 columns)

# Adding 1D arrays as additional columns
df['stimulus'] = data['stimulus'].flatten()
df['repetition'] = data['repetition'].flatten()
df['restimulus'] = data['restimulus'].flatten()
df['rerepetition'] = data['rerepetition'].flatten()

# Convert 'glove' (22 columns) to DataFrame and merge
df_glove = pd.DataFrame(data['glove'], columns=[f'glove_{i+1}' for i in range(data['glove'].shape[1])])
df = pd.concat([df, df_glove], axis=1)
```

## Executando YData-Profiling

```python
profile = ProfileReport(df, title="DB1 - S1_A1_E1 Report")

output_folder = "../reports"
output_file = os.path.join(output_folder, "DB1_S1_A1_E1_Report.html")

# Create the folder if it does not exist
os.makedirs(output_folder, exist_ok=True)

# Export the report to an HTML file
profile.to_file(output_file)
```
